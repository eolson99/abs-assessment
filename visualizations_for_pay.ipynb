First, I make a histogram of average pay in the tech/pro/science sector from state to state.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pay_df = pd.read_csv('csvs/pay_data.csv')
AvgPay = pay_df[['AvgPay']]

ax = sns.histplot(AvgPay).set_title('Average Pay in Scientific, Technical, Professional Occupations by State')
plt.xlabel('Average Annual Pay in Thousands of Dollars')
plt.ylabel('Number of States')
plt.legend().remove()
```

Next, I make a horizontal bar chart of the top 5 highest paying states.

```python
top_5 = pay_df.sort_values(by=['AvgPay'], ascending=False).head(5)[['State', 'AvgPay']]
ax2 = top_5.plot(title='Top 5 Highest Paying States for Scientific, Technical, Professional Occupations', kind='barh', x='State', y='AvgPay')
ax2.set_xlabel('Averge Annual Pay in Thousands of Dollars')
ax2.invert_yaxis()
ax2.get_legend().remove()
```
