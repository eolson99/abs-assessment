```python
import pandas as pd
sales = pd.read_csv('ai_sell.csv')
sales.columns
```

```python
sales.dtypes
```

```python
sales = sales[['State','PercentFirmsSellingAI']]
```

```python
AIUseChart = sales.sort_values('PercentFirmsSellingAI', ascending = False).head(10)
chart = AIUseChart.plot.bar(x = 'State', y = 'PercentFirmsSellingAI', title = "Percentage of Firms that Sell AI")
chart.set_ylabel("% of sales")
```

```python
cloudsales = pd.read_csv('cloud_sell.csv')
cloudsales.columns

CloudChart = cloudsales.sort_values("PercentFirmsSellingCloud", ascending = False).head(10)

chart = CloudChart.plot.bar(x = 'State',y = 'PercentFirmsSellingCloud', title = 'Percentage of Firms that Sell Cloud Technology')
chart.set_ylabel("% of firms")
```
